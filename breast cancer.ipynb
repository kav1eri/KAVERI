```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Assuming your DataFrame is named df
# If not, replace df with your actual DataFrame variable

# Get the list of column names excluding 'id' and 'diagnosis'
columns_to_plot = df.columns[2:]

# Set up subplots
fig, axes = plt.subplots(nrows=len(columns_to_plot)//3 + 1, ncols=3, figsize=(15, 20))

# Flatten the axes for easy iteration
axes = axes.flatten()

# Loop through each column and create a normal distribution curve
for i, column in enumerate(columns_to_plot):
    ax = axes[i]
    sns.histplot(df[column], kde=True, ax=ax, color='skyblue', stat='density')
    
    # Generate points for normal distribution curve
    mean = df[column].mean()
    std_dev = df[column].std()
    x = np.linspace(mean - 3*std_dev, mean + 3*std_dev, 100)
    y = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp( - (x - mean)**2 / (2 * std_dev**2) )
    ax.plot(x, y, color='red')
    
    ax.set_title(f'Distribution of {column}')
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.grid(True, linestyle='--', alpha=0.7)

# Remove empty plots
for i in range(len(columns_to_plot), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
import pandas as pd

# Assuming your DataFrame is named df
# If not, replace df with your actual DataFrame variable
# Select the variables you're interested in
selected_variables = ['radius_mean', 'area_mean', 'perimeter_mean', 'texture_mean']

# Create a new DataFrame with only the selected variables
selected_df = df[selected_variables]

# Define a function to remove outliers using IQR
def remove_outliers_iqr(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_data = data[(data >= lower_bound) & (data <= upper_bound)]
    return filtered_data, Q1, Q3, IQR

# Apply the function to remove outliers for each selected variable
for variable in selected_variables:
    filtered_data, Q1, Q3, IQR = remove_outliers_iqr(selected_df[variable])
    selected_df[variable] = filtered_data

    # Print quartile values
    print(f'{variable}: Q1={Q1}, Q3={Q3}, IQR={IQR}')

# Replace the original columns in the main DataFrame with the cleaned ones
df[selected_variables] = selected_df[selected_variables]
```

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Assuming your DataFrame is named df
# If not, replace df with your actual DataFrame variable

# Select the variables you're interested in
selected_variables = ['texture_mean', 'radius_mean', 'area_mean', 'perimeter_mean']

# Create a new DataFrame with only the selected variables
X = df[selected_variables]

# Define the target variable
y = df['diagnosis']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the imputer
imputer = SimpleImputer(strategy='mean')

# Fit and transform the imputer on the training data
X_train_imputed = imputer.fit_transform(X_train)

# Transform the testing data using the trained imputer
X_test_imputed = imputer.transform(X_test)

# Initialize the Logistic Regression model
model = LogisticRegression(random_state=42)

# Train the model on the training data
model.fit(X_train_imputed, y_train)

# Make predictions on the testing data
y_pred = model.predict(X_test_imputed)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grid for Logistic Regression
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}

# Initialize Logistic Regression model
model = LogisticRegression(random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')

# Fit GridSearchCV on imputed training data
grid_search.fit(X_train_imputed, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print(f'Best Hyperparameters: {best_params}')

# Make predictions on the testing data using the best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test_imputed)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy with Best Model: {accuracy:.2f}')
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

y_prob = best_model.predict_proba(X_test_imputed)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```
